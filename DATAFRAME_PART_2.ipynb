```
columns = ["language","users_count"]
data = [("Java", "20000"), ("Python", "100000"), ("Scala", "3000"),("Golang",None),("Unix",None),(None,'1000'),(None,"30000")]

sc=spark.sparkContext
rdd=sc.parallelize(data)
df=rdd.toDF()
print(df.printSchema())

df1=rdd.toDF(columns)
print(df1.printSchema())

df1.filter(df1.users_count.isNull()).show()
df1.filter(df1['users_count'].isNull()).show()
df1.filter(col('users_Count').isNull()).show()
```
